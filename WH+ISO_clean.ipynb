```python
#Bringing in CSVs use in data compilation 
import pandas as pd
from sqlalchemy import create_engine
import psycopg2 
import io

```

```python
#Reading in the World Happiness Report from 2021
happiness = pd.read_csv("Data/world-happiness-report-2021.csv")
happiness.head()
```

```python
#Reading in the ISO codes for each country
isocc = pd.read_csv("Data/iso-country-codes.csv")
isocc.head()
```

```python
#Combining the World Happiness Report & the ISOs 
Happiness_ISO = pd.merge(left=happiness,right=isocc, left_on='Country name', right_on='Country name')
Happiness_ISO.shape
Happiness_ISO.head()
```

```python
#Dropping the 2 code, ISO#, & numberic codes since those aren't needed to combine with our data

Happiness_ISO.drop(['Regional indicator','Alpha-2 code', 'Numeric code','ISO 3166-2','Standard error of ladder score','upperwhisker','lowerwhisker'], axis='columns', inplace=True)
Happiness_ISO.head()
```

```python
#change code to be more descriptive
Happiness_ISO = Happiness_ISO.rename(columns={'Alpha-3 code': 'country_code'})
Happiness_ISO.head()
```

```python
#create dataframe to load countries table
countries_df=Happiness_ISO.drop(['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'], axis='columns')
countries_df.rename(columns={'Country name':'country_name'}, inplace=True)
countries_df.head()
```

```python
# Create csv to use in WB-Data-Retrieval to drop countries not include in Happiness Index
countries_df.to_csv('countires.csv')
```

```python
#DataFrame to load happiness index into sql
Happiness_Index=Happiness_ISO.drop(['Country name'], axis='columns')
Happiness_Index.rename(columns={'Dystopia + residual':'Dystopia_residual','Explained by: Log GDP per capita':'Explained by Log GDP per capita','Explained by: Social support':'Explained by Social support','Explained by: Healthy life expectancy':'Explained by Healthy life expectancy','Explained by: Freedom to make life choices':'Explained by Freedom to make life choices','Explained by: Generosity':'Explained by Generosity','Explained by: Perceptions of corruption':'Explained by Perceptions of corruption'}, inplace=True)
Happiness_Index.head()
```
