```python
#Bringing in CSVs use in data compilation 
import pandas as pd
```

```python
#Reading in the World Happiness Report from 2021
happiness = pd.read_csv("Data/world-happiness-report-2021.csv")
happiness
```

```python
#Reading in the ISO codes for each country
isocc = pd.read_csv("Data/iso-country-codes.csv")
isocc
```

```python
#Combining the World Happiness Report & the ISOs 
Happiness_ISO = pd.merge(left=happiness,right=isocc, left_on='Country name', right_on='Country name')
Happiness_ISO.shape
Happiness_ISO
```

```python
#Dropping the 2 code, ISO#, & numberic codes since those aren't needed to combine with our data

Happiness_ISO.drop(['Alpha-2 code', 'Numeric code','ISO 3166-2'], axis='columns', inplace=True)
Happiness_ISO
```

```python
#During discussions we decided that we didn't need all of the Happiness Report data along with the alpha-3 because we are using it for joining table. Ugly & inefficient, but we go there.
Happiness_ISO.drop(['Regional indicator', 'Ladder score', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'], axis='columns', inplace=True)
Happiness_ISO
```

```python
#A CSV of the original Happiness Report + ISO was created called Happiness_Report_ISO but that won't be used. 
Happiness_ISO.to_csv('Country_with_ISO.csv')
```
